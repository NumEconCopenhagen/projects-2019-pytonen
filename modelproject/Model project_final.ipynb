```python
# We start by importing the relevant packages
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy import optimize
import sympy as sm
sm.init_printing(use_unicode=True)
```

## Visualization of the Consumers Utility

The problem considers the consumer's utility of two goods:

\\[ u(x_1,x_2) = x_1 x_2 + 2 x_2 \\]

The consumption-possibility frontier is \\[(x_1,x_2) \in (0,\infty)\times(0,\infty)\\] Also, we assume that \\[p_1,p_2,I>0\\]

Firstly we visualize the consumer's utility

```python
# We define the utility function
def u(x1,x2):
    u = x1*x2 + 2 * x2
    return u
```

```python
# We construct 2 vectors, x1 and x2, within the interval 0 to 10.
x1_vec = np.linspace(0, 10, 40)
x2_vec = np.linspace(0, 10, 40)

# We then construct a grid of the two vectors constructed above.
x1_grid, x2_grid = np.meshgrid(x1_vec, x2_vec)

# Within this grid we consider the utility function
u_grid=u(x1_grid, x2_grid)

# We plot the graph in 3d
fig = plt.figure()
ax = fig.add_subplot(1,1,1,projection='3d')
ax.plot_surface(x1_grid,x2_grid,u_grid, cmap=plt.cm.jet, cstride=1, rstride=1)

# Then we add labels and invert the axes
ax.set_xlabel('$x_1$')

ax.set_ylabel('$x_2$')
ax.set_zlabel('$u$')
ax.invert_xaxis()
```

## Solving the Consumer's Utility Maximization Problem

We consider the following utility maximization problem: 

\\[ 
\begin{eqnarray*}
\max_{x_{1},x_{2}} x_{1} x_{2} + 2 x_{2}\ \text{u.b.}\ p_{1} x_{1} + p_{2} x_{2} = I
\end{eqnarray*}
\\]  

We define our utility and budget constraint in terms of sympy such that we can computate symbolic expressions with the variables in question.

```python
# Converting the variables into symbolic expression
x1 = sm.symbols('x_1')
x2 = sm.symbols('x_2')
p1 = sm.symbols('p_1')
p2 = sm.symbols('p_2')
I  = sm.symbols('I')
U = sm.symbols('U')

# Constructing the utility function and budget constraint in terms of symbolic expressions
utility = (x1*x2 + 2*x2)
budget_constraint = sm.Eq(p1*x1 + p2*x2, I)
```

We wish to find the consumers demand for $x_1$ and $x_2$ respectivly. First we isolate $x_2$

```python
# Isolate x2 withinthe budget constraint
x2constraint = sm.solve(budget_constraint, x2)
x2constraint[0]
```

Then we substitute our result for $x_2$ into the the utility function

```python
utility_subx2 = utility.subs(x2, x2constraint[0])
utility_subx2
```

Now we are able to maximize this equation by finding the deriviate with respect to $x_1$

```python
focx1 = sm.diff(utility_subx2, x1)
focx1
```

We set this equal to zero and solve for $x_1$

```python
x1sol = sm.solve(sm.Eq(0,focx1), x1)
x1sol[0]
```

We substitute the result above for $x_1$ into our budget constraint expressed as a function for $x_2$

```python
x1constraint = sm.solve(budget_constraint, x1)
utility_subx1 = utility.subs(x1, x1constraint[0])
focx2 = sm.diff(utility_subx1, x2)
x2sol = sm.solve(sm.Eq(focx2, 0), x2)
x2sol[0]
```

Note: The above solutions are applicable for $I > 2p_1$. We choose to disregard the corner solutions in the situation where $ I \leq 2p_1$.

## A Specific Case

We let prices $p_1 = 4$ and $p_2 = 1$ and $I = 16$. Due to a change in circumstances for the production of the two goods, a fall in the price of good 1 falls to $p_1 = 1$

```python
# We define the result above for x1 and x2 in terms of functions instead of symbolic expressions to calculate the solution
def x2func(p1,p2,I):
    return (I/2 + p1)/p2

def x1func(p1,p2,I):
    return (I/2 - p1)/p1
```

```python
# We consider x1 and x2 for the case of p1=4, p2=1 and I=16
print('x2 =',x2func(4,1,16))
print('x1 =',x1func(4,1,16))
print('utility =', u(x1func(4,1,16),x2func(4,1,16)))
```

```python
# We then consider x1 and x2 for the case of p1=1, p2=1 and I=16
print('x2 =',x2func(1,1,16))
print('x1 =',x1func(1,1,16))
print('utility =', u(x1func(1,1,16),x2func(1,1,16)))
```

**We show the results graphically**

```python
# Redefining utility function as an equivalence, such that we are able to isolate x2
utility_Eq = sm.Eq(U,utility)

# We isolate x2 in the utility function and budget constraint such that we are able to plot it in the same graph
print('x2(utility) =', sm.solve(utility_Eq,x2))
print('x2(budget) =', sm.solve(budget_constraint, x2))
```

With pretty writing: <br/>

$x_2^{budget} = \frac{I - p_1 x_1}{p_2} $ <br/>
$x_2^{utility} = \frac{U}{x_1 + 2} $

```python
# We define the indifference curve and budget constraint in terms of x2 by functions such that we can plot these
def budget(x1,I,p1,p2):
    return (I-p1*x1)/p2

def indiff(x1,u,p1,p2):
    return u/(x1+2)

# We plot the indifference curve, budget constraint and optimum for p1=4
fig = plt.figure(dpi=100)
ax = fig.add_subplot(1,1,1)
ax.plot(x1_vec,budget(x1_vec,16,4,1),label='Budget Constraint') 
ax.plot(x1_vec,indiff(x1_vec,36,4,1),label='Indifference Curve')
ax.scatter(x1func(4,1,16),x2func(4,1,16),u(x1func(4,1,16),x2func(4,1,16)),color='black',label='Optimum')

# We add labels and a title
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
plt.title('Optimal consumption bundle given $p_1 = 4$, $p_2 = 1$ and $I = 16$')
ax.legend();

# We plot the indifference curve, budget constraint and optimum for p1=1
fig = plt.figure(dpi=100)
ax = fig.add_subplot(1,1,1)
ax.plot(x1_vec,budget(x1_vec,16,1,1),label='Budget Constraint')
ax.plot(x1_vec,indiff(x1_vec,81,1,1),label='Indifference Curve')
ax.scatter(x1func(1,1,16),x2func(1,1,16),u(x1func(1,1,16),x2func(1,1,16)),color='black',label='Optimum')

# We add labels and a title
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
plt.title('Optimal consumption bundle for $p_1=1$, $p_2 = 1$ and $I = 16$')
ax.legend();
```

## Interactive Solution

To visualize the optimal consumption bundle in terms of difference prices and incomes, we plot an interactive chart, where budget constraint, indifference curve and optimum depend on prices and incomes. Also we plot two indifference curves above and below the optimal curve to visualize how the indifference curves change.

```python
# We define the plot in terms of a function to be able to use a slider to change the independent variables
def _figure(x1_vec, I, p2, p1):
    fig=plt.figure(dpi=60)
    ax = fig.add_subplot(1, 1, 1)
    x1_vec = np.linspace(0, 2.3*x1func(p1, p2, I), 40)
    x2_vec = np.linspace(0, 2.3*x2func(p1, p2, I), 40)
    ax.plot(x1_vec,budget(x1_vec,I,p1,p2),label="Budget Constraint")
    x1 = x1func(p1,p2,I) # We define x1 as a function in terms of the independent varibles
    x2 = x2func(p1,p2,I) # We define x2 as a function in terms of the independent varibles   
    
    # Given the definitions of x1 and x2, we are now able to define the utility function in terms of the independent varibles
    u_plot = x1*x2 + 2*x2 
    ax.plot(x1_vec,indiff(x1_vec,u_plot,p1,p2),label="Indifference Curve in optimum")
    ax.plot(x1_vec,indiff(x1_vec,u_plot*0.5,p1,p2),label="Indifference Curve x0.5")
    ax.plot(x1_vec,indiff(x1_vec,u_plot*1.5,p1,p2),label="Indifference Curve x1.5")
    ax.scatter(x1,x2,color='black',label='Optimum')
    plt.title('Optimal Consumption Bundle')
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')
    ax.legend();
    plt.grid(which = 'both')

# Then we plot the sliders of p1, p2 and I
import ipywidgets as widgets
widgets.interact(_figure, 
                x1_vec = widgets.fixed(x1_vec), 
                I = widgets.FloatSlider(description="$I$", min=0.1, max=50, step=0.01, value=16),
                p2 = widgets.FloatSlider(description="$p_2$", min=0.01, max=16, step=0.01, value=1),
                p1 = widgets.FloatSlider(description="$p_1$", min=0.01, max=16, step=0.01, value=1));
```

We notice that when $I \leq 2p_1$, the optimal consumption bundle impossibly requires a negative value for $x_1$ confirming the existence of a corner solution.

## Alternative Solution by Numerical Analysis

Above we chose to solve the problem symbolically using sympy, but we could also have chosen to solve it numerically using scipy. We perform the numerical analysis below to answer the same problem.

```python
# We find the optimal consumption bundle for the following values
I = 16
p1 = 1
p2 = 1

# We redefine our utility function
def u_new(x1,x2):
    return x1*x2 + 2 * x2

# We define our indifference curve with a negative sign to use a minimizer to maximize utility
def func(x1,I,p1,p2):
    x2 = (I-p1*x1)/p2
    return -u_new(x1,x2)

# We call the scipy solver
solution = optimize.minimize_scalar(
    func,method='bounded',
    bounds=(0,I/p2),args=(I,p1,p2))


# We unpack the solution
x1 = solution.x
x2 = (I-p1*x1)/p2
u_func = u_new(x1,x2)

# Then we print our solution for p1=1, p2=1 and I=16
print(f'x1 = {x1:.1f}')
print(f'x2 = {x2:.1f}')
print(f'u  = {u_func:.1f}')
print(f'I-p1*x1-p2*x2 = {I-p1*x1-p2*x2:.2f}') 
```

By this we conclude that the same result is found by numerical analysis as found by the analysis performed earlier:

```python
print('x1 =',x1func(1,1,16))
print('x2 =',x2func(1,1,16))
print('u =', u(x1func(1,1,16), x2func(1,1,16)))
print('I-p1*x1-p2*x2 =', 16 - 1*x1func(1,1,16) - 1*x2func(1,1,16))
```

We thus show that consumption amounts given the prices utilises the entire income.

# Social Planner

We now assume that prices for both goods are dropped. Instead a social planner has to distribute the goods that consumer A bought when $p_1 = 1$ and $p_2 = 1$ between consumer A and a newly entered consumer B. This means that the social planner has to distribute a total of $x_1 = 7$ and $x_2 = 9$ between the consumers. We assume no endowments <br/> 
Consumer A's utility function remains the same while consumer B has the utility function: <br/>
$ u_B = 3 \frac{1}{8}* x_{1B} + 3 \frac{1}{8} * x_{2B} $ <br/>
The social utility function is given by $W(u_A, u_B) = min(u_A, u_B)$ meaning the social planner seeks to maximise the utility for the consumer with the lowest utility. The solution will therefore give them both equal utility.

## Algebraic Solution for $W(u_A, u_B) = min(u_A, u_B)$

To solve this problem algebraically, we begin by defining the required symbols:

```python
X1   = sm.symbols('X1')   # Total amount of good 1
X2   = sm.symbols('X2')   # Total amount of good 2
```

```python
utilityA      = X1*X2 + 2*X2  # Utility function for consumer A
utilityB      = 3.125*X1 + 3.125*X2  # Utility function for consumer B
```

```python
# To optimize the social planners problem we calculate the first order conditions for each consumer w.r.t. both goods
foc1utilityA = sm.diff(utilityA, X1)
foc2utilityA = sm.diff(utilityA, X2)
foc1utilityB = sm.diff(utilityB, X1)
foc2utilityB = sm.diff(utilityB, X2)
```

```python
# We begin by calculating MRS for consumer B since its marginal utilities are equal.
mrsB = - foc1utilityB / foc2utilityB
mrsB
```

```python
# Next we calculate MRS for consumer A
mrsA = - foc1utilityA / foc2utilityA
mrsA
```

Then we set the MRS for consumer A equal to the MRS for consumer B to find the line along which consumer A has a marginal rate of substitution equal to -1. 

```python
# setting MRs for consumer A and B equal to eachother
mrseq = sm.Eq(mrsB,mrsA)

# Solving for consumer A, hence isolating x2
x2_demand_A = sm.solve(mrseq,X2)
x2_demand_A[0]
```

We see that as long as $x_2 = x_1 + 2$, consumer A has a MRS equal to $-1$, and along this line the solutions will all be pareto optimal for the consumers, but not for the social planner. <br/>
Next we find the utility of consumer A expressed by $x_1$ only by inserting $x_1 + 2$ in $x_2$'s place

```python
# substituting the expression for x2 found above into the utlitity function for consumer A
uA_x1 = utilityA.subs(X2,x2_demand_A[0])
uA_x1
```

We define the amounts of good 1 and 2 for consumer B to be the excess amounts from consumer A, and then replace $x_2$ with $x_1 + 2$ in the same way as for consumer A:

```python
# Defining excess of goods in terms of x1 as found above, to find the optimal level of goods for consumer B
excess_x1 = 7 - X1
excess_x2 = 9 - x2_demand_A[0]

# Substituting these into the utility functions for consumer B
uB_x1 = utilityB.subs(X1,excess_x1)
uB_x2 = uB_x1.subs(X2,excess_x2)
uB_x2
```

We set utility functions for both consumers expressed as a function of $x_1$ only equal to each other, and express this equality in the form of a second degree equation. This will give us a value of $x_1$ where both consumers' utility are equal.

```python
uA_uB = sm.Eq(uA_x1-uB_x2,0)
uA_uB
```

```python
# Solving optimal level of x1 for consumer A
x1_solution_A_ = sm.solve(uA_uB,X1)
x1_solution_A_
```

We disregard the negative value $-13.25$ as the goods are only defined in the non-negative numbers and therefore cannot have a negative value. This gives us the solution:

```python
x1_solution_A=x1_solution_A_[1]
x1_solution_A
```

The solution where the social planner has optimized its utility function $W(u_A, u_B) = min(u_A, u_B)$ thus gives consumer A: $x_1 = 3$. <br/>
We insert the solution for $x_1$ into the expression for $x_2$ to find the optimal level of $x_2$:

```python
# Substituting the solution into the expression for x2
x2_solution_A = x2_demand_A[0].subs(X1,x1_solution_A)
x2_solution_A
```

We find that the solution planner optimal solution with $x_1 = 3$ gives a corresponding value of $x_2 = 5$. This also follows from the function $ x_2 = x_1 + 2$ that describes the ling along which the MRS for consumer A is equal to $-1$. <br/>
Now, we're able to solve for consumer B:

```python
# Inserting good 1 solution for consumer A into excess function for consumer B
x1_solution_B = excess_x1.subs(X1,x1_solution_A)
x1_solution_B
```

```python
# Inserting solution for consumer A into excess function for consumer B
x2_solution_B_ = excess_x2.subs(x2_demand_A[0],x2_solution_A) # substituting consumers A's demand for x2 into excess function for consumer B
x2_solution_B = x2_solution_B_.subs(X1,x1_solution_A)# substituting consumers A's demand for x1 into solution function for consumer B
x2_solution_B
```

The optimal solution for the social planner thus has both $x_1 = 4$ and $x_2 = 4$ for consumer B. <br/>
Then we find the utility for the 2 consumers:

```python
utilityA_sol = x1_solution_A*x2_solution_A + 2*x2_solution_A
utilityB_sol = 3.125*x1_solution_B + 3.125*x2_solution_B

print(f'utility A = {utilityA_sol:.1f}')
print(f'utility B = {utilityB_sol:.1f}')
```

Since both consumers at this point has a marginal rate of substitution equal to -1 while both of their utilities are $u_A(3, 5) = u_B(4, 4) = 25$, this distribution of goods is the solution to maximising $W(u_A, u_B) = min(u_A, u_B)$. 

## Plot

```python
X1_total = 7
X2_total = 9

# We use three functions with x2 isolated to plot them in the same diagram.
def x2_A(x1):
    return utilityA_sol/(x1+2)

def x2_MRS(x1):
    return x1 + 2

def x2_B(x1):
    return (utilityB_sol-3.125*(x1))/3.125


x1 = np.linspace(0,8,100)

plt.plot(x1,x2_A(x1))
plt.plot(x1,x2_B(x1))
plt.plot(x1,x2_MRS(x1),color='black')
plt.legend(['Consumer A','Consumer B','MRS=1'],loc=1)
plt.axis([0,X1_total,0,X2_total]);
plt.ylabel('$x_2$, Consumer A')
plt.xlabel('$x_1$, Consumer A')
plt.title('Edgeworth box')
plt.grid();
```

As depicted above we have a pareto optimal distribution of goods, when consumer A has $3$ of good 1 and $5$ of good 2. Consumer B receives $4$ of each goods, which corresponds to the secondary axis system with origin in (7,9), where the axes are unnumbered.  <br/> 
We see that in the point where the utility functions for consumers A and B crosses the black line, the marginal rates of substitution are equal to $-1$.
